```python
plt.figure(figsize=(12,8))
sns.boxplot(data=data, orient='h')
plt.title('Outliers Detection')
plt.show()
```

```python
q1 = data['MonthlyIncome'].quantile(0.25)
q3 = data['MonthlyIncome'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
data = data[ (data['MonthlyIncome'] >= lower_bound) & (data['MonthlyIncome'] <= upper_bound)]
        
```

```python
plt.figure(figsize=(12,8))
fig = data.boxplot(column='MonthlyIncome')
fig.set_title('Checking for Outliers in Salary')
fig.set_ylabel('Monthly Salary')
```

```python
# Encoding Target Variable

data['Attrition'] = data['Attrition'].replace('Yes', 1)

data['Attrition'] = data['Attrition'].replace('No', 0)

data.head()
```

```python
data.corr()
```

```python
drop_columns = ['EmployeeCount', 'StandardHours']

data = data.drop(columns = drop_columns)
    
sns.heatmap(data.corr(), cmap = 'icefire', linewidths = 1)
plt.title('Understanding Correlation with Visualization')
plt.show()
```

```python
# Converting float data to integer data
 
data['NumCompaniesWorked'] = data['NumCompaniesWorked'].round().astype(int)
     
data['TotalWorkingYears'] = data['TotalWorkingYears'].round().astype(int)
     
data.info()     
```

```python
cat_columns = data.select_dtypes(include=['object']).columns

cat_columns
```

```python
data = pd.get_dummies(data, columns = cat_columns)

data.info()
```

```python
data.head()
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(data.corr(), cmap = 'icefire', linewidths = 0.5)
plt.title('Correlation Plot')
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
```

```python
Y = data['Attrition']

X = data.drop(['Attrition'], axis = 1)

# # k_best = SelectKBest(score_func = f_classif, k=2)

# # X = k_best.fit_tranform()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(solver = 'liblinear', random_state=0)

log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)

y_pred
```

```python
log_reg.predict_proba(X_test)[:,0]
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
coefficients = log_reg_100.coef_[0]

# Compute the absolute values of the coefficients
feature_importance = np.abs(coefficients)

# Create a DataFrame to display feature importance
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importance
})

# Sort the features by importance in descending order
importance_df = importance_df.sort_values(by='Importance', ascending=True).reset_index(drop=True)

# Plot feature importance
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance in Logistic Regression')
plt.show()

print(importance_df)
```
